```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
corona_dataset_csv=pd.read_csv("C:/Users/INNVERSE/Desktop/project/covid19_data.csv")
corona_dataset_csv.head()
```

```python
corona_dataset_modified=corona_dataset_csv.drop(["Lat","Long"], axis=1)
corona_dataset_modified.head(15)
```

```python
corona_dataset_aggregate=corona_dataset_modified.groupby("Country/Region").sum()
corona_dataset_aggregate.head(10)
```

```python
corona_dataset_modified.shape
```

```python
corona_dataset_aggregate.shape
```

```python
corona_dataset_aggregate.loc["China"].plot()
corona_dataset_aggregate.loc["Italy"].plot()
corona_dataset_aggregate.loc["Spain"].plot()
corona_dataset_aggregate.loc["India"].plot()
plt.legend()
```

```python
corona_dataset_aggregate.loc["China"].diff().plot()
corona_dataset_aggregate.loc["Italy"].diff().plot()
plt.legend()
```

```python
countries=list(corona_dataset_aggregate.index)
max_inf_rate=[]
for c in countries :
    max_inf_rate.append(corona_dataset_aggregate.loc[c].max())
corona_dataset_aggregate["max_inf_rate"]=max_inf_rate 
corona_dataset_aggregate.head()
```

```python
corona_data=pd.DataFrame(corona_dataset_aggregate["max_inf_rate"])
corona_data.head(10)
```

```python
happiness_report_csv=pd.read_csv("C:/Users/INNVERSE/Desktop/project/worldhappiness_report.csv")
happiness_report_csv.head()
```

```python
useless_col=["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report=happiness_report_csv.drop(useless_col,axis=1)
happiness_report.head()
```

```python
happiness_report.set_index('Country or region', inplace = True)
```
